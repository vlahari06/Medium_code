```
import requests
from bs4 import BeautifulSoup
```

```
res = requests.get('https://en.wikipedia.org/wiki/Harry_Potter_(film_series)')
```

```
content = res.text
```

```
content
```

```
soup=BeautifulSoup(content,'html.parser')
```

```
soup
```

```
req_table = soup.find('table', {'class':'wikitable'})
```

```
req_table
```

```
movie_list = []
```

```
for tr in req_table.find_all('tr')[1:]:
    tds = tr.find_all('td')
    movie_list.append(tds[1].text)
```

```
movie_list
```

```
mylist = []
```

```
for tr in req_table.find_all('tr')[1:]:
    tds = tr.find_all('td')
    name = tds[1].contents[0]
    ##print(tds)
    ##print(name)
    ##print("---")
    a = name.find('a')
    mylist.append(a.get('title'))   
```
